Realizar una gráfica que permita ver el porcentaje de municipios por estado con índices de marginación "muy bajo", "bajo", "medio", "alto" y "muy alto". Guarda la gráfica en archivo png

```python
categorias = {
    'MB': 'Muy bajo',
    'B': 'Bajo',
    'M': 'Medio',
    'A': 'Alto',
    'MA': 'Muy alto'
}

municipios_por_estado = datos.groupby(['CVE_ENT', 'NOM_ENT', 'GM_2020']).size().unstack(fill_value=0)

porcentaje_por_estado = municipios_por_estado.apply(lambda row: row / row.sum() * 100, axis=1)

colores = ['#2ecc71', '#3498db', '#f1c40f', '#e67e22', '#e74c3c']

plt.figure(figsize=(10, 8))
porcentaje_por_estado.plot(kind='bar', stacked=True, color=colores)
plt.title('Porcentaje de municipios por estado con diferentes niveles de marginación')
plt.xlabel('Estado')
plt.ylabel('Porcentaje')
plt.xticks(range(len(porcentaje_por_estado.index)), porcentaje_por_estado.index)
leyenda = [plt.Rectangle((0, 0), 1, 1, color=colores[i]) for i in range(len(colores))]
plt.legend(leyenda, categorias.values(), title='Categoría de Marginación', bbox_to_anchor=(1.01, 1), loc='upper left')
plt.savefig('porcentaje_marginacion.png')
plt.show()
```

Grafica la relación de porcentaje de analfabetismo respecto al porcentaje de poblaciones en localidades de menos de 5,000 habitantes.

```python
localities_5000 = datos[datos['POB_TOT'] < 5000]


percentage_analfabetismo = localities_5000['ANALF']
percentage_poblacion_5000 = localities_5000['POB_TOT'] / localities_5000['POB_TOT'].sum() * 100

plt.figure(figsize=(10, 8))
plt.scatter(percentage_poblacion_5000, percentage_analfabetismo, color='blue', alpha=0.5)
plt.title('Relación de Porcentaje de Analfabetismo y Porcentaje de Población en Localidades < 5000 Habitantes')
plt.xlabel('Porcentaje de Población en Localidades < 5000 Habitantes')
plt.ylabel('Porcentaje de Analfabetismo')
plt.show()
```

Desarrolla un nuevo DataFrame con indicadores interesantes por estado que se obtengan de los datos a nivel municipal y que pudieran ser de importancia a la hora de definir políticas públicas. Se va a calificar la originalidad de los indicadores. Justifica las decisiones que tomes, y guarda el nuevo dataframe en formato parquet.

```python
indicadores_por_estado = datos.groupby(['CVE_ENT', 'NOM_ENT']).agg({
    'POB_TOT': 'sum',  
    'ANALF': 'mean',  
    'SBASC': 'mean',  
    'IM_2020': 'mean',  
    'VHAC': 'mean',  
})

indicadores_por_estado.rename(columns={
    'POB_TOT': 'Población Total',
    'ANALF': 'Porcentaje de Analfabetismo',
    'SBASC': 'Porcentaje de Población sin Acceso a Servicios Básicos',
    'IM_2020': 'Índice de Marginación',
    'VHAC': 'Valor de Viviendas con Acceso a Agua Entubada',
}, inplace=True)

indicadores_por_estado.to_parquet('indicadores_por_estado.parquet')
```
